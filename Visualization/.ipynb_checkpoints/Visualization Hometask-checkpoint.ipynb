Столбец **`published`** (время публикации) содержит строки. Чтобы мы могли работать с этими данными как с датой/временем публикации, приведём их к типу `datetime`:

```python
print(df.published.dtype)
df['published'] = pd.to_datetime(df.published, yearfirst=True)
print(df.published.dtype)
```

Создадим несколько столбцов на основе данных о времени публикации:

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]

df['dayofweek'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
```

Определим новую переменную, показывающиую, в какое время дня была осуществлена публикация:

```python
def times_of_day(x):
    if x >= 12 and x < 18:
        return 'afternoon'
    elif x >= 18 and x <= 23:
        return 'evening'
    elif x >= 0 and x <= 4:
        return 'night'
    return 'morning'
```

```python
df['times_of_day'] = df['hour'].apply(times_of_day)
```

Посмотрим на среднее количество комментариев под постом в зависимости от времени дня:

```python
sns.barplot(x='times_of_day', y='comments', data=df)
plt.title('Average number of comments for each time of the day', size=20)
plt.ylabel('number of comments')
plt.xlabel('time of the day');
```

В среднем, под постами, опубликованными в ночное время суток, оставляют больше комментариев.

## 5\. Сравните субботы и понедельники

Правда ли, что по субботам авторы пишут в основном днём, а по понедельникам — в основном вечером?

```python
last_two = df[df.dayofweek.isin([6, 1])].sort_values(by='dayofweek')
last_two['dayofweek'] = last_two['dayofweek'].replace({6: 'saturday', 1: 'monday'})

fig = plt.figure(figsize=(20, 10))
sns.countplot(x='dayofweek', data=last_two, hue='times_of_day', palette='Set3')
plt.title('Number of posts on Monday and Saturday depending on time of the day')
plt.xlabel('day of week')
plt.ylabel('number of posts');
```
